```python
#André Paiva 2C
#PERGUNTA: Uma alta demanda energética e um alto índice urbano podem indicar um alto nível de desenvolvimento?

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

e=pd.read_excel("energy_use_per_person.xlsx")
u=pd.read_excel("urban_population_percent_of_total.xlsx")
idh=pd.read_excel("hdi_human_development_index.xlsx")
```

```python
e_2010=e.loc[:,['Países',2010]]
u_2010=u.loc[:,['Países',2010]]
idh_2010=idh.loc[:,['Países',2010]]
```

```python
e_idx=e_2010.set_index('Países')
u_idx=u_2010.set_index('Países')
idh_idx=idh_2010.set_index('Países')
```

```python
dv1=idh_idx.join(e_idx, how='inner', lsuffix='_idh', rsuffix='_e')
dv2=dv1.join(u_idx,how='right',rsuffix='_u')
dv1["2010_e_log"] = np.log(dv1["2010_e"])
dv1.rename(columns={dv1.columns[0]:'IDH'},inplace=True)
dv1.rename(columns={dv1.columns[1]:'Eletricity Use per Person'},inplace=True)
dv2.rename(columns={dv2.columns[0]:'IDH'},inplace=True)
dv2.rename(columns={dv2.columns[1]:'Eletricity Use per Person'},inplace=True)
dv2.rename(columns={dv2.columns[2]:'Urban Population(%)'},inplace=True)
dv1
```

```python
dv1.plot.scatter(x="2010_e_log", y =  "IDH")
```

```python
dv2 = idh_idx.join(u_idx, how='inner', lsuffix='_idh', rsuffix='_u')
dv2.plot.scatter(x="Urban Population(%)", y =  "2010_idh", loglog=False)
```

```python
def reta(dados_x, dados_y):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
 
    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
 
    fig = plt.figure(figsize=(8, 6))
    plt.scatter(dados_x, dados_y, c='red', alpha=0.8)
    plt.title('x versus y')
    plt.xlabel('eixo x')
    plt.ylabel('eixo y')
 
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color='blue')
 
    plt.tight_layout()
    plt.show()
```

```python
reta(dv1["2010_e_log"], dv1["IDH"],)
```
