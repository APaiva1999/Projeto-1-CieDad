**PERGUNTA**: Uma alta demanda energética e um alto índice urbano podem indicar um alto nível de desenvolvimento?</p>
<p>*Introdução*<br>Nos dias de hoje, em que temos um mundo tão globalizado, é normal associar que países com grande IDH tenham uma sociedade bem estruturada, com cidades planejadas e grandes investimentos em infraestrutura, industria etc. Contudo, os países com maior IDH tem grandes níveis de urbanização ou é possível ter uma ótima estrutura no campo? Além disso, o gasto energético que essas cidades geram, somadas ao gasto industrial entre outros, podemos afirmar que o uso de eletricidade e os índices de urbanização estão correlacionados ao IDH? </p>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

e=pd.read_excel("energy_use_per_person.xlsx")
u=pd.read_excel("urban_population_percent_of_total.xlsx")
idh=pd.read_excel("hdi_human_development_index.xlsx")


def reta(dados_x, dados_y, nome_x,  nome_y):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre ',nome_x,'  e ',nome_y,':')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)

    fig = plt.figure(figsize=(8, 6))
    plt.scatter(dados_x, dados_y, c='red', alpha=0.8)
    plt.title(nome_x + " versus "+ nome_y)
    plt.xlabel(nome_x)
    plt.ylabel(nome_y)

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color='blue')

    plt.tight_layout()
    plt.show()
```

```python
e_2010=e.loc[:,['Países',2010]]
u_2010=u.loc[:,['Países',2010]]
idh_2010=idh.loc[:,['Países',2010]]
```

```python
e_idx=e_2010.set_index('Países')
u_idx=u_2010.set_index('Países')
idh_idx=idh_2010.set_index('Países')
```

```python
dv1=idh_idx.join(e_idx, how='inner', lsuffix='_idh', rsuffix='_e')
dv2=dv1.join(u_idx,how='right',rsuffix='_u')
dv1["2010_e_log"] = np.log(dv1["2010_e"])
dv2.rename(columns={dv2.columns[2]:"2010_u"},inplace=True)
dv1["2010_u_log"]= np.log(dv2["2010_u"])
dv2["2010_e_log"]= dv1["2010_e_log"]
dv2["2010_u_log"]= dv1["2010_u_log"]
dv2
```

```python
dv1.plot.scatter(x="2010_e_log", y =  "2010_idh")
```

```python
dv2 = idh_idx.join(u_idx, how='inner', lsuffix='_idh', rsuffix='_u')
dv2.plot.scatter(x="2010_u", y =  "2010_idh", loglog=False)
```

**Análise**<p>
Observando a tabela dv2 e os gráficos acima, é possível notar que países mais urbanizados tendem a ter um IDH maior e também que países em que as pessoas tem um alto gasto energético tem um IDH bom, apesar de não aparentar uma correlação tão gritante com IDH quanto a urbanização, como observa-se nos casos do Bahrain, que possui um IDH de 0.812, mas um gasto energético de 10200 e uma índice de urbanização de 87.50 se comparado ao da Australia que possui um IDH de 0.927, mas com quase metade do gasto energétic, cerca de 5800, e com uma taxa de urbanização um pouco superior de cerca de 88.70% .

```python
reta(dv1["2010_e_log"], dv1["2010_idh"], "Uso Energético/Pessoa (log)", "IDH")
```

```python
reta(dv1["2010_u_log"], dv1["2010_idh"], "Índice de Urbanização(%)","IDH")
```

```python
dv2.rename(columns={dv2.columns[0]:'IDH'},inplace=True)
dv2.rename(columns={dv2.columns[1]:'Eletricity Use per Person'},inplace=True)
dv2.rename(columns={dv2.columns[2]:'Urban Population(%)'},inplace=True)
dv2.rename(columns={dv2.columns[3]:'Eletricity Use per Person (log)'},inplace=True)
dv2.rename(columns={dv2.columns[4]:'Urban Population (log)'}, inplace=True)
dv2
```

**Conclusão**<p>
Sendo assim, conclui-se que pelos gráficos de reta e das tabelas acima, que há correlações entre o IDH e Uso Energético/pessoa em log, com uma correlação próxima de 0.846 e também entre IDH e População Urbana em %,com uma correlação bem pouco abaixo do 0.700, mas que há uma correlação que é visível através da análise do gráfico de IDH x Urbanização que possui alguns outliers (como Barbados que tem um IDH de 0.780 com um índice de urbanização de apenas 32.20%) que alteram esta correlação, mas com um bom número de países próximos ao que seria a reta. Desta Forma, podemos afirmar, sim, que é possível relacionar alta demanda energética e urbanização com um IDH alto, mas existem alguns países fora do padrão, estes que são minoria.
